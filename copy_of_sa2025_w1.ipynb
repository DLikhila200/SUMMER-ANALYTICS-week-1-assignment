### 1. What is name of car that has the highest horsepower?  

```python
# Find maximum horsepower value
max_hp = df['horsepower'].max()

# Find the car that has this horsepower
car_with_max_hp = df[df['horsepower'] == max_hp].index

# Print the car name
print(car_with_max_hp.values)
```

### 2. How many cars have mpg ≥ 35?

```python
count_mpg_35 = (df['mpg'] >= 35).sum()
print(count_mpg_35)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Select cars with hp > 100 and weight < 3000
selected_cars = df[(df['horsepower']) > 100 & (df['weight'] < 3000)]

# Find the most common origin among selected cars
most_common_origin = selected_cars['origin'].mode()[0]

print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_acceleration_japan = round(df[df['origin'] == 'japan']['acceleration'].mean(), 2)
print(mean_acceleration_japan)
```

### 5. Which year had the highest average mpg?

```python
# Calulate the average mpg for each model year
avgmpg_peryear = df.groupby('model_year')['mpg'].mean()

# Find the year with highest average mpg
highest_avg_mpg_year = avgmpg_peryear.idxmax()

print(highest_avg_mpg_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Step 1: Calculate the median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with mpg greater than median
high_mpg_cars = df[df['mpg'] > median_mpg].copy()

# Step 3: Calculate horsepower to weight ratio
high_mpg_cars['hp_to_weight'] = high_mpg_cars['horsepower'] / high_mpg_cars['weight']

# Step 4: Find the maximum ratio
max_ratio = high_mpg_cars['hp_to_weight'].max()

# Step 5: Get car(s) with the max ratio
best_cars = high_mpg_cars[high_mpg_cars['hp_to_weight'] == max_ratio]

# Step 6: Print the car name(s)
best_cars = best_cars.reset_index()
print(best_cars[['name', 'horsepower', 'weight', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Group the data by year and origin and calculate average mpg
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

# Plotting
plt.figure(figsize=(12, 6))

# Plot each origin as a separate line
for origin in avg_mpg_by_year_origin.columns:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin], label=origin)

# Add a title and axis labels to the plot
plt.title('Average mpg over the years by origin')
plt.xlabel('Model year')
plt.ylabel('Average mpg')

# Add a legend to differentiate lines by origin
plt.legend(title='Origin')

# Add a grid for better readability
plt.grid(True)

# Adjust layout to prevent clipping
plt.tight_layout()

# Display the plot
plt.show()


```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Add meaningful plot titles and axis titles.

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Step 1: Inspect column names and reset index so 'name' definitely becomes a column
print("Original columns:", df.columns.tolist())
df = df.reset_index(drop=True)  # ensures nothing is stuck in the index

# Step 2: Standardize the car-name column to 'name'
if 'index' in df.columns and df['index'].dtype == object:
    df.rename(columns={'index': 'name'}, inplace=True)
elif 'level_0' in df.columns and df['level_0'].dtype == object:
    df.rename(columns={'level_0': 'name'}, inplace=True)

print("'name':", df.columns.tolist())

# Step 3: Standardize the model_year column
if 'model year' in df.columns:
    df.rename(columns={'model year': 'model_year'}, inplace=True)
    print("'model_year':", df.columns.tolist())
else:
    print("'model_year' not found")
# Step 4: Double-check we now have exactly 'name', 'model_year', and 'mpg'
print("Final columns before grouping:", df.columns.tolist())

# Step 5: Group by 'name' and compute:
#   - appearances = how many distinct model_year values per car name
#   - mpg_std    = standard deviation of mpg across those years
#   - mpg_mean   = average mpg across those years
stats = df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std     =('mpg',    'std'),
    mpg_mean    =('mpg',    'mean')
).reset_index()

# Step 6: Filter to only those cars with at least 2 appearances and low variation (std < 1.0)
consistent_cars = stats[
    (stats['appearances'] >= 2) &
    (stats['mpg_std']      <  1.0)
]

# Step 7: Sort by number of appearances (descending), then by average mpg (descending)
consistent_cars_sorted = consistent_cars.sort_values(
    by=['appearances', 'mpg_mean'],
    ascending=[False,         False]
)

# Step 8: Print the final result: model name, number of appearances, and average mpg
print(consistent_cars_sorted[['name', 'appearances', 'mpg_mean', 'mpg_std']])
```
